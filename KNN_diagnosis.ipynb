```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn import preprocessing
from sklearn.preprocessing import scale
import sklearn as sk
```

```python
data = pd.read_csv('E:\\New Volume\\Downloads\\wbcd.csv')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
X = data.loc[:,['radius_mean','area_mean','smoothness_mean']]
y = data.loc[:,'diagnosis']

```

```python
X.head()
```

```python
y.head()
```

```python
X = scale(X)
X = pd.DataFrame(X)
#X.head()
type(X)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30)
```

```python
X_train
```

```python
model_knn = KNeighborsClassifier(n_neighbors = 20, metric='euclidean')

model_knn.fit(X_train,y_train)
```

```python
pv = model_knn.predict(X_test)
np.mean(pv == y_test)
```

```python
accuracy_score(pv,y_test)
confusion_matrix(y_test,pv)
```
